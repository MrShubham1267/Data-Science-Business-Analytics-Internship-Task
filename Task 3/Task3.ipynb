# Exploratory Data Analysis

```python
sns.pairplot(df)
```

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue="Region",palette='bwr')
```

```python
df['Ship Mode'].unique()
```

```python
seg=df.Segment.value_counts().reset_index()
seg.columns=('Segment','Count')
seg
```

```python
plt.pie(x="Count",labels="Segment",data=seg,radius=2,autopct="%.2f",pctdistance=0.4)
```

```python
plt.pie(df['Ship Mode'].value_counts(),labels=df['Ship Mode'].value_counts().index)
plt.title('Ship mode')
plt.show()
```

```python
plt.title("Region")
df['Region'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True);
```

```python
plt.title("Segment")
df['Segment'].value_counts().plot.pie(autopct ='% 1.1f %%', shadow = True);
```

```python
plt.pie(df['Category'].value_counts(),labels= df['Category'].value_counts().index)
plt.title("Category")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('CATEGORIES VS REGION')
sns.countplot(x=df['Category'],hue=df['Region'],palette='Set3')
plt.xticks()
```

```python
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(9,5))
plt.title('CATEGORIES VS REGION')
sns.countplot(x=df['Category'],hue=df['Region'],palette='Set3')
plt.xticks()
```

```python
high_sale=df[df["Sales"]>3000]
high_sale.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=high_sale["Sales"],y=high_sale["State"],data=df)
```

```python
df.Sales.max()
```

```python
df.Sales.min()
```

```python
low_profit=df[df["Profit"]<1]
low_profit.head()
```

```python
plt.figure(figsize=(9,5))
low_sale=df[df["Sales"]<1]
sns.barplot(y=low_sale["State"],x=low_sale["Sales"],data=low_sale,palette='rainbow')
```

```python
data=['Sales','Quantity','Discount','Profit']
plt.figure(figsize=(10,5))
sns.heatmap(df[data].corr(),annot=True, fmt='.4f',center=0);
```

From the above heatmap we can see that Sales and Profit have high positive correlation.Quantity and Discount have almost no correlation ,Profit and Discount are negatively correlated
